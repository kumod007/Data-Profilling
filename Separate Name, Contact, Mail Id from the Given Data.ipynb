### Separating Contact Using Details Given in DataFrame.

```python
contact = []

for i in df["Details"]:

    x = i
    for j in range(len(i)):
        if i[j].isalpha():
            x = x.replace(i[j],",")
            
    x = x.replace("@","")
    x = x.replace(".","")
    y = x.split(",")
    z = sorted(y,reverse=True)
    #print(z)
    
    for i in z:
        if len(i)==10:
            if (i[0]=="9") or (i[0]=="8") or (i[0]=="7"):
                #print(i)
                contact.append(i)
                break
            else:
                #print("No Contact")
                contact.append("No Contact")
                break
        else:
            #print("No Contact")
            contact.append("No Contact")
            break
```

### Adding Contact Column in The Datafreame.

```python
df["Contact"] = contact
```

```python
mail = []

for i in df["Details"]:
    for j in df["Contact"]:
        if j in i:
            i = i.replace(j,"")
            mail.append(i)
    if i not in mail:   
        mail.append(i)
    
        
            
```

```python
name = []

for i in df["Mail Id"]:
    a = i.split("@")
    b = a[0]
    
    for j in b:
        if j.isdigit():
            b = b.replace(j,"")
    if len(b)>1:
        name.append(b.title())
    else:
        name.append("No Name")
```
