### Creating Function to get row index having DateTime Data Type.

```python
list1 = []

for index,value in enumerate(df["Data"]):
    if type(value)==str:
        continue
    else:
        list1.append(index)
```

```python
print("Indices of rows containing date-time data type values are:\n")
print(list1)
```

### Dropping All Those Rows We Obtained.

### Creating a function to get rows which are required.

* We Need only those rows having **CR, SALE, Customer, **BNG** as values.
* So using a function we will extract those rows index values.

```python
list2 = []

for index,value in enumerate(df["Data"]):
    if "BNG" in value:
        list2.append(index)
    elif "CR" in value:
        list2.append(index)
    elif "SALE" in value:
        list2.append(index)
    elif "Customer" in value:
        list2.append(index)
```

```python
print("Required indices of rows are:\n")
print(list2)
```

**Since We have got those useful rows index. But we have to drop those rows which are not present in our list2. So we will create one more function to get those rows.**

```python
list3 = []

for i in range(len(df)):
    if i not in list2:
        list3.append(i)
```

```python
print("Indices of rows which are not required are:\n")
print(list3)
```

### Dropping those rows which are not required.

```python
df.drop(list3,axis=0,inplace=True)
```

**Observation**
* We can observe that in every Invoice bill there are **TKT NO** which we have to extract and create new column as equal to the tickets numbers.

## Separating Ticket No From Those Rows.

```python
ticket_data = []

for i in df["Data"]:
    a=i.split("TKT No-")
    if len(a)==1:
        ticket_data.append(" ")
    if (len(a)==2):
        ticket_data.append(a[1])
```

**Creating Ticket Data new column**

```python
df["Ticket_Data"]=ticket_data
```

### Analzing Ticket_Data Column

```python
a =[]

for i in df["Ticket_Data"]:
    a.append(len(i.split(",")))
```

```python
print("Minimum Tickets is:",min(a))
print("Maximum Tickets is:",max(a))
```

**Since there are maximum 6 tickets numbers so we have to create 6 new columns such as Ticket No 1, Ticket No 2, Ticket No 3, Ticket No 4, Ticket No 5, Ticket No 6**

**Checking How many rows are having ticket no 6 times**

```python
for i in df["Ticket_Data"]:
    a=i.split(",")
    if len(a)==6:
        print(a)
```

### Creating Function To Separate All Those Different Ticket Numbers.

```python
ticket1 = []
ticket2 = []
ticket3 = []
ticket4 = []
ticket5 = []
ticket6 = []

for i in df["Ticket_Data"]:
    a = i.split(",")
    if len(a)==1:
        ticket1.append(" ")
        ticket2.append(" ")
        ticket3.append(" ")
        ticket4.append(" ")
        ticket5.append(" ")
        ticket6.append(" ")

    elif len(a)==2:
        ticket1.append(a[0])
        ticket2.append(" ")
        ticket3.append(" ")
        ticket4.append(" ")
        ticket5.append(" ")
        ticket6.append(" ")
    
    
    elif len(a)==3:
        ticket1.append(a[0])
        ticket2.append(a[1])
        ticket3.append(" ")
        ticket4.append(" ")
        ticket5.append(" ")
        ticket6.append(" ")
    
    elif len(a)==4:
        ticket1.append(a[0])
        ticket2.append(a[1])
        ticket3.append(a[2])
        ticket4.append(" ")
        ticket5.append(" ")
        ticket6.append(" ")
    
    elif len(a)==5:
        ticket1.append(a[0])
        ticket2.append(a[1])
        ticket3.append(a[2])
        ticket4.append(a[3])
        ticket5.append(" ")
        ticket6.append(" ")
        
    elif len(a)==6:
        ticket1.append(a[0])
        ticket2.append(a[1])
        ticket3.append(a[2])
        ticket4.append(a[3])
        ticket5.append(a[4])
        ticket6.append(a[5])
```
